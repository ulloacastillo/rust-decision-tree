```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
iris = pd.read_csv('iris.csv', header=None)
```

```python
with open("iris_txt.txt", "w") as f:
    data = iris.iloc[:, :4].values.flatten().tolist()
    f.write(str(data))
```

```python
train = iris.iloc[38:, :]
test = iris.iloc[:38, :]

plt.hist(train[4], 5, label='train', align='mid')
plt.hist(test[4], 5, label='test', align='mid')
plt.legend()


tree = DecisionTreeClassifier(splitter='best')

tree.fit(train.iloc[:, :-1], train.iloc[:, -1])

yhat = tree.predict(test.iloc[:, :-1])

print(accuracy_score(test.iloc[:, -1], yhat))
print(confusion_matrix(yhat, test.iloc[:, -1] ))
```
